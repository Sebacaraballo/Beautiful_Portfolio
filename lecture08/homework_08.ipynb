We are tasked with assessing the usefulness of a tuberculosis test.
The prior information I is:

> The percentage of the population infected by tuberculosis is 0.4\%.
We have run several experiments and determined that:
+ If a tested patient has the disease, then 80\% of the time the test comes out positive.
+ If a tested patient does not have the disease, then 90\% of the time the test comes out negative.

To facilitate your analysis, consider the following logical sentences concerning a patient:

> A: The patient is tested and the test is positive.

> B: The patient has tuberculosis.

A. Find the probability that the patient has tuberculosis (before looking at the result of the test), i.e., $p(B|I)$. This is known as the base rate or the prior probability.

**Answer:**


The prior probability is the likelihood of having the disease before any test is performed. Based on the population data provided (0.4%):$$p(B|I) = 0.004$$

B. Find the probability that the test is positive given that the patient has tuberculosis, i.e., $p(A|B,I)$.

**Answer:**

This is the Sensitivity of the test. You stated that if a patient has the disease, the test is positive 80% of the time:$$p(A|B,I) = 0.80$$

C. Find the probability that the test is positive given that the patient does not have tuberculosis, i.e., $p(A|\neg B, I)$.

**Answer:**

First, we look at the Specificity (True Negative rate), which is 90% or $p(\neg A|\neg B,I) = 0.90$.The probability of a False Positive is the complement of the specificity:$$p(A|\neg B,I) = 1 - p(\neg A|\neg B,I) = 1 - 0.90 = 0.10$$

D. Find the probability that a patient that tested positive has tuberculosis, i.e., $p(B|A,I)$.

**Answer:**

We use Bayes' Theorem to find the probability that a person who tested positive actually has the disease:$$p(B|A,I) = \frac{p(A|B,I)p(B|I)}{p(A|I)}$$First, calculate the total probability of a positive test $p(A|I)$:$$p(A|I) = p(A|B,I)p(B|I) + p(A|\neg B,I)p(\neg B|I)$$$$p(A|I) = (0.80 \times 0.004) + (0.10 \times 0.996) = 0.0032 + 0.0996 = 0.1028$$Now, plug it back into Bayes' Theorem:$$p(B|A,I) = \frac{0.0032}{0.1028} \approx 0.0311 \text{ (or 3.11%)}$$

E. Find the probability that a patient that tested negative has tuberculosis, i.e., $p(B|\neg A, I)$. Does the test change our prior state of knowledge about about the patient? Is the test useful?

**Answer:**

We calculate the probability of having the disease despite a negative result:$$p(B|\neg A,I) = \frac{p(\neg A|B,I)p(B|I)}{p(\neg A|I)}$$$$p(\neg A|B,I) = 1 - 0.80 = 0.20$$$$p(\neg A|I) = 1 - p(A|I) = 1 - 0.1028 = 0.8972$$$$p(B|\neg A,I) = \frac{0.20 \times 0.004}{0.8972} \approx 0.00089 \text{ (or 0.089%)}$$Analysis:The test does change our knowledge. It moved our probability from 0.4% to either ~3.1% (positive) or ~0.09% (negative). However, even with a positive result, there is only a 3% chance the patient has TB. This is because the disease is so rare that the "false alarms" from the 99.6% healthy population outweigh the "true hits" from the 0.4% sick population.

F. What would a good test look like? Find values for

$$
p(A|B,I)= p(\text{test is positive} |\text{has tuberculosis},I),
$$

and

$$
p(A| \neg B, I) = p(\text{test is positive}|\text{does not have tuberculosis}, I),
$$

so that

$$
p(B|A, I) = p(\text{has tuberculosis}|\text{test is positive}, I) = 0.99.
$$

There are more than one solutions. How would you pick a good one? Thinking in this way can help you set goals if you work in R\&D. If you have time, try to figure out whether or not there exists such an accurate test for tuberculosis

**Answer:**

To achieve $p(B|A,I) = 0.99$, we need to solve:$$0.99 = \frac{p(A|B,I) \times 0.004}{p(A|B,I) \times 0.004 + p(A|\neg B,I) \times 0.996}$$If we assume a perfect sensitivity ($p(A|B,I) = 1.0$):$$0.99 = \frac{0.004}{0.004 + p(A|\neg B,I) \times 0.996}$$$$0.99(0.004 + 0.996x) = 0.004 \implies 0.00396 + 0.98604x = 0.004$$$$0.98604x = 0.00004 \implies x \approx 0.0000405$$Conclusion:To be 99% sure of a diagnosis in this population, your False Positive rate $p(A|\neg B,I)$ must be roughly 0.004% (which means a specificity of 99.996%). In R&D, you would prioritize reducing false positives over slightly increasing sensitivity, as the low base rate makes the test extremely sensitive to even tiny error rates in the healthy population.
